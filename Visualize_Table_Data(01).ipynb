```
import pandas as pd
```

```
df = pd.read_csv('https://raw.github.com/pandas-dev/pandas/master/pandas/tests/io/data/csv/iris.csv')
df
```

# Parall coodinate

```
pd.plotting.parallel_coordinates(df, 'Name')
```

ปรับ paramiter color

```
pd.plotting.parallel_coordinates(df, 'Name', color=('#552627' , '#4ECDC4' , '#C7F464')) #html color
```

```
pd.plotting.parallel_coordinates(df, 'Name', color=('r' , 'g' , 'b')) #ตัวย่อของสี  r g b
```

```
pd.plotting.parallel_coordinates(df, 'Name', color=((1,0,0) , (0,1,0) , (0,0,1))) # rgb value (r,g,b)
```

```
pd.plotting.parallel_coordinates(df, 'Name', color=((1,0.8,0) , (0,0.7,0) , (0,0,1)))
```

```
pd.plotting.parallel_coordinates(df, 'Name', color=((1,0,0) , (0,1,0) , (0.6,0,1)))
```

```
pd.plotting.parallel_coordinates(df, 'Name', colormap='cool')
```

```
pd.plotting.parallel_coordinates(df, 'Name', colormap='tab10')
```

# เริ่ม Matplotlib

```
from matplotlib import pyplot as plt
```

###แปลง data จาก pandas dataframe เป็น numpy array

```
df.values
```

```
df[['PetalWidth','PetalLength', 'SepalWidth']]
```

```
np_data=df.iloc[:,:-1].to_numpy()
np_data
```

##visualize array

```
import matplotlib
```

```
matplotlib.rcParams['figure.figsize']=[15,15] #กำหนดขนาดความกว้างความยาวของตัวมี่พล็อต
plt.imshow(np_data[:10,:],interpolation='nearest')
```

การใช้ subplot

```
plt.subplot(1,2,1)
plt.imshow(np_data[:50,:],interpolation='nearest')

plt.subplot(1,2,2)
plt.imshow(np_data[50:100,:],interpolation='nearest')

plt.subplot(1,3,3)
plt.imshow(np_data[150:100,:],interpolation='nearest')
```
